```python
import random
import cv2
import torch
import numpy as np
from PIL import Image
from diffusers import ControlNetModel, StableDiffusionXLControlNetPipeline, AutoencoderKL
from diffusers import EulerAncestralDiscreteScheduler
from controlnet_aux import HEDdetector
import gradio as gr
```

```python
controlnet = ControlNetModel.from_pretrained(
    "xinsir/controlnet-scribble-sdxl-1.0", torch_dtype=torch.float16
)
vae = AutoencoderKL.from_pretrained("madebyollin/sdxl-vae-fp16-fix", torch_dtype=torch.float16)
scheduler = EulerAncestralDiscreteScheduler.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0", subfolder="scheduler"
)
pipe = StableDiffusionXLControlNetPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    controlnet=controlnet,
    vae=vae,
    safety_checker=None,
    torch_dtype=torch.float16,
    scheduler=scheduler,
).to("cuda")
```

```python
def generate_scribble_from_image(image_path):
    """
    Generates a scribble representation from an image using a HED detector.
    """
    image = cv2.imread(image_path)
    processor = HEDdetector.from_pretrained("lllyasviel/Annotators")
    controlnet_img = processor(image, scribble=False)
    controlnet_img = np.array(controlnet_img)
    controlnet_img = cv2.GaussianBlur(controlnet_img, (0, 0), 3)

    # Simulate human sketching
    random_val = int(round(random.uniform(0.01, 0.10), 2) * 255)
    controlnet_img[controlnet_img > random_val] = 255
    controlnet_img[controlnet_img < 255] = 0
    return Image.fromarray(controlnet_img)


def generate_image(sketch, user_prompt, selected_style, input_type, image=None):
    """
    Generates an image based on input type (sketch or image), style, and user prompt.
    """
    print("Starting image generation...")

    # Check input type and process accordingly
    if input_type == "Sketch":
        controlnet_img = np.array(sketch["composite"])
        controlnet_img = controlnet_img.astype(np.uint8)
        controlnet_img = cv2.cvtColor(controlnet_img, cv2.COLOR_RGB2GRAY)
        controlnet_img[controlnet_img > 127] = 255
        controlnet_img[controlnet_img <= 127] = 0
        controlnet_img = Image.fromarray(controlnet_img)
    elif input_type == "Image" and image:
        controlnet_img = generate_scribble_from_image(image)
    else:
        return None  # Return None if neither input is valid

    # Resize to fit model input size (1024x1024)
    width, height = controlnet_img.size
    ratio = np.sqrt(1024.0 * 1024.0 / (width * height))
    new_width, new_height = int(width * ratio), int(height * ratio)
    controlnet_img = controlnet_img.resize((new_width, new_height))

    # Retrieve style-specific prompts and combine with user prompt
    style_data = next(style for style in style_list if style["name"] == selected_style)
    style_prompt = style_data["prompt"]
    negative_prompt = style_data["negative_prompt"]
    final_prompt = f"{user_prompt}, {style_prompt}"

    # Generate image using the pipeline
    try:
        images = pipe(
            prompt=final_prompt,
            negative_prompt=negative_prompt,
            image=controlnet_img,
            controlnet_conditioning_scale=1.0,
            guidance_scale=7.5,
            width=new_width,
            height=new_height,
            num_inference_steps=50,
        ).images

        print("Image generation complete.")
        return images[0] if images else None
    except Exception as e:
        print(f"Error during image generation: {e}")
        return None
```

```python
def gradio_interface():
    with gr.Blocks() as demo:
        gr.Markdown("# Sketch or Image-to-Art Generator")

        # Input type selector
        input_type = gr.Radio(
            choices=["Sketch", "Image"], label="Choose Input Type", value="Sketch"
        )

        # Inputs
        sketch_input = gr.Sketchpad(label="Draw your sketch", visible=True)
        image_input = gr.Image(type="filepath", label="Upload your image", visible=False)
        user_prompt = gr.Textbox(label="Enter your prompt", placeholder="Describe the scene or object")
        style_dropdown = gr.Dropdown(
            label="Select an art style",
            choices=[style["name"] for style in style_list],
            value=style_list[0]["name"],
        )

        # Output
        output_image = gr.Image(label="Generated Image")

        # Toggle function for input type
        def toggle_input(input_type):
            if input_type == "Sketch":
                return gr.update(visible=True), gr.update(visible=False)
            else:
                return gr.update(visible=False), gr.update(visible=True)

        # Update visibility based on input type
        input_type.change(toggle_input, inputs=[input_type], outputs=[sketch_input, image_input])

        # Generate button
        generate_button = gr.Button("Generate Image")
        generate_button.click(
            generate_image,
            inputs=[sketch_input, user_prompt, style_dropdown, input_type, image_input],
            outputs=[output_image],
        )

    demo.launch(debug=True)

gradio_interface()
```
